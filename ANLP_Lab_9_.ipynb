## **Task 3:** QA system handles questions that cannot be answered.

I did follow the instructions on the lab file "CAI451-Lab9.pdf"

**Citation and all rights for the *code*  reserved to Professor *Badr Aldaihani***

```python
def main():
    # Load the question-answering pipeline
    qa_pipeline = pipeline("question-answering", model="distilbert/distilbert-base-cased-distilled-squad")

    print("Welcome to the CAI451-Question Answering System!")
    print("Type 'exit' to quit.")

    # Context for the questions
    context = """Princess Nourah bint Abdulrahman University (PNU) aspires to be a beacon of knowledge and values for women.
    In 2006, a royal decree was issued that established the first university for girls in Riyadh.
    PNU is the largest women's university in the world.
    PNU is one of the successes of the care and attention that has been directed toward women's higher education.
    Currently, Her Excellency Prof. Al-Eisa serves as the President of Princess Nourah bint Abdulrahman University."""

    while True:
        user_question = input("You: ").strip()

        if user_question.lower() == 'exit':
            print("Goodbye!")
            break

        try:
            result = qa_pipeline(question=user_question, context=context)
            answer = result['answer']
            confidence = result['score']

            # Use a confidence threshold to decide if the answer is reliable
            if confidence < 0.5:
                print("System: Sorry, I don't know the answer.")
            else:
                print("System:", answer)
        except Exception as e:
            print("System: Sorry, I don't know the answer.")

if __name__ == "__main__":
    main()
```

## **Task 4:** create Chatbot (google translator)

```python
def translation_chatbot():
    print("Welcome to the CAI451 English-to-Arabic translation chatbot!")
    print("Type 'exit' to exit the conversation.\n")

    while True:
        # Get user input (English text)
        english_text = input("You: ").strip()

        # Exit condition
        if english_text.lower() == 'exit':
            print("Chatbot: Goodbye!")
            break

        try:
            # Translate the text to Arabic
            arabic_translation = GoogleTranslator(source='en', target='ar').translate(english_text)
            print(f"Chatbot: {arabic_translation}")
        except Exception as e:
            print(f"Chatbot: Sorry, an error occurred: {e}")

# Run the chatbot
translation_chatbot()
```

## **Task 5:** create Chatbot (LLMbased)

I did follow the instructions on the lab file "CAI451-Lab9.pdf"

**Citation and all rights for the *code*  reserved to Professor *Badr Aldaihani***

```python
def translation_chatbot():
    print("Welcome to the CAI451 English-to-Arabic translation chatbot powered by an LLM!")
    print("Type 'exit' to stop the conversation.\n")

    while True:
        # Get user input (English text)
        english_text = input("You: ")

        # Exit condition
        if english_text.lower() == 'exit':
            print("Chatbot: Goodbye!")
            break

        # Use GPT to translate the text to Arabic
        try:
            response = openai.ChatCompletion.create(
                model="gpt-3.5-turbo",  # Use a valid model ID
                messages=[
                    {"role": "user", "content": f"Translate the following English text to Arabic:\n\n{english_text}"}
                ],
                max_tokens=60,  # Limit the response length
                temperature=0.3  # Lower temperature for more focused translations
            )

            # Extract and print the translation
            translation = response.choices[0].message['content'].strip()
            print(f"Chatbot: {translation}")

        except Exception as e:
            print(f"Chatbot: Sorry, an error occurred: {e}")

# Run the chatbot
translation_chatbot()
```
